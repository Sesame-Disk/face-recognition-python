```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
import imshowpair

backends = ["opencv", "ssd", "dlib", "mtcnn", "retinaface", "mediapipe"]
models = [
    "VGG-Face","Facenet","Facenet512","OpenFace","DeepFace","DeepID","ArcFace","Dlib",
]
img1 = plt.imread("/content/taylor2.jpg")
img2 = plt.imread("/content/taylor5.jpg")
verification = DeepFace.verify(
    img1_path=img1, img2_path="/content/taylor5.jpg", model_name= models[2], detector_backend=backends[2]
)
imshowpair.imshowpair(img1, img2)
print(verification)
```
